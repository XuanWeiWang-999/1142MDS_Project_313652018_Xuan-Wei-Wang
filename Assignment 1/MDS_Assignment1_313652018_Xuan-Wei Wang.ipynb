# 1. 五階段構析 Five Stage Analytics

## 1.1 Wafer Map Detective Problem - 晶圓缺陷檢測分類 問題與情境
該問題通常發生於**半導體晶圓製造產線（如蝕刻或微影製程）**。在先進製程中，晶圓圖 (Wafer Map) 的瑕疵數據往往差異極大，但有些具備特定致命瑕疵（如 Scratch 刮痕、Edge-Ring 邊緣環狀），這導致傳統的機器學習分類器難以收斂，容易產生誤報 (False Negative)，讓瑕疵品流入下一道工序，可能因此造成工廠巨大的成本損失。  
  
藉此，希望導入基於去噪分數匹配 (Denoising Score Matching) 等機制的生成式 AI 來擴增少數瑕疵類別，能平衡數據集，可以大幅提升後續預測模型的準確度與產線良率。

## 1.2 Five Stage Analytics

首先奠基於**敘述性分析 (Descriptive)**。透過蒐集產線上的感測器數據與自動光學檢測 (AOI) 的晶圓圖 (Wafer Map)，系統能即時監控並具象化目前的良率狀態。當敘述性指標呈現出特定的異常分佈（例如發現極少數但致命的刮痕或邊緣環狀瑕疵）時，我們便可以沿著異常分布與警訊往成品上游找到問題製造者，以及造成這些特定圖譜的根本原因，流程自然推進至**診斷性分析 (Diagnostic)**，例如特定蝕刻機台的氣體流量失衡。  
  
找到異常造成的根本原因後，我們可以試著設計模型，根據及時串流數據提前預測一批晶圓是否有高機率報廢，這促使流程進入**預測性分析 (Predictive)**。比如我們可導入擴散模型 (Diffusion Models) 來生成高品質的瑕疵影像以擴增數據集。這份經過強化的數據被用來訓練機器學習分類器，使其能夠在機台發生微小偏移時，不僅能提前「預測」未來產出的晶圓是否面臨高報廢風險，也能洞見其異常週期。  
  
若系統偵測到該批晶圓有高機率報廢，肯定也相對應要考慮報廢晶圓的成本，比如停機作業損失與回收成本等；或是回頭向前一步驟調整參數等。因此數據流必須無縫接軌至**處方性分析 (Prescriptive)**，也就是工廠人員需設計決策出一套系統流程，使其不僅會顯示良率下降的預測警告，還會綜合考量停機檢修的成本與晶圓報廢的損失，開出最佳「處方」——例如建議立刻暫停某機台，或是建議微調特定製程參數。  
  
最後，當處方性系統的建議被驗證為高度可靠時，這條決策鏈的終極目標便是走向 **自主性分析 (Autonomous)** ────也就是系統會將人類決策者從迴圈中移除，建立先進製程控制 (APC) 的閉環系統，使得機台能夠根據預測與處方結果，即時且自主地進行參數補償與修正。而這些自動修正後的生產結果，又會產生新的數據，重新回饋到最初的「敘述性分析」階段進行監控，達成無人化、具備自我修正能力的閉環控制系統。

* **預測性分析（Predictive）**：故障特徵預測CNN 模型能夠對未見過的晶圓進行推理（Inference），預測其所屬的缺陷類別與發生機率 。

## 1.7 Critical Thinking

我認為比較特別的一點是，因為瑕疵晶圓其實佔總產量是較低的比例，甚至稀缺，所以樣本數往往不夠支撐這樣的研究。所以為了生成樣本，他們使用 *Poisson point process*生成：$$P(k,\Lambda)=\frac{\Lambda^n}{k!}e^{-\Lambda}$$其中$\Lambda$這個參數決定了一個區段裡平均有多少瑕疵晶圓的比例，$k$則是事件的數量。但後來有了Generative AI的發展，這樣傳統的方式反而變得特別了。  
  
不過在真實工廠端，其實狀況還是比較多，Poisson point process的方法其實也有限。比如若真實製程中的雜訊或瑕疵型態(邊緣效應、化學反應導致的特殊分布)，用該模型就會在表現上打折扣。也許可以針對製程中某種特定機台去模擬其物理特性導致的特定瑕疵聚集產生樣本，雖然參數或過程可能會更複雜，但相信表現也會相較於只用數學方法來得好。或是，利用Generative AI，無論Diffusion model或是flow matching都可以，加入物理特性去生成樣本，會更為方便。

# 2. Manufactoring System Analysis

## 2.1 產線資訊
產出率 $TH$ 為每單位時間內生產的合格產品數量，所以我們可以得到：
| 工作站編號 | 平行等效機台數 ($m$) | 加工時間 ($t$) | 產能 $TH_i$ (個/小時) |
| :---: | :---: | :---: | :---: |
| 1 | 3 | 5 | $3/5 = 0.6$ |
| 2 | 3 | 8 | $3/8 = 0.375$ |
| 3 | 6 | 12 | $6/12 = 0.5$ |
| 4 | 2 | 4 | $2/4 = 0.5$ |
| 5 | 5 | 12 | $5/12 \approx 0.417$ |

由上表可知，第 **2** 站產能最小，所以為瓶頸站。  
* 瓶頸站產出率 ($r_b$)：$r_b = 0.375$ (個/小時)
* 最小生產週期時間 ($T_0$)：即總加工時間（Raw Process Time, RPT）。
  * $T_0 = 5 + 8 + 12 + 4 + 12 = \mathbf{41}$ (小時)
* 關鍵在製品水準 ($W_0$)：指系統在達到最大產出率且無等待時間下的最小 WIP 數量。
  * $W_0 = r_b \times T_0 = 0.375 \times 41 = \mathbf{15.375}$ (個)
* 產線穩定條件：投料速度（Release Rate, $R$）必須小於瓶頸站的產出率。
  * 穩定條件：$R < r_b$ (即 $R < 0.375$)

## 2.2 最佳績效
* 生產週期時間(CT)：$$CT(w) = \begin{cases} 41, & \text{if } w \leq 15.375 \\ w/0.375, & \text{otherwise.} \end{cases}$$  
* 產出率(TH)：$$TH(w) = \begin{cases} w/41, & \text{if } w \leq 15.375 \\ 0.375, & \text{otherwise.} \end{cases}$$

## 2.3 產線模擬分析
我們將繪製兩種圖以表現三者關係：WIP與TH; WIP與CT

```python
import numpy as np
import matplotlib.pyplot as plt

m = [3, 3, 6, 2, 5]          # 平行機台數
t = [5, 8, 12, 4, 12]        # 加工時間
th_stations = [m[i]/t[i] for i in range(len(m))]

rb = min(th_stations)        
T0 = sum(t)                  
W0 = rb * T0                 

w_values = np.linspace(1, 50, 100)

th_best = np.array([min(w/T0, rb) for w in w_values])
ct_best = np.array([max(T0, w/rb) for w in w_values])

th_pwc = (w_values / (W0 + w_values - 1)) * rb
ct_pwc = T0 + (w_values - 1) / rb

```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# WIP vs Throughput (TH)
ax1.plot(w_values, th_best, label='Best Case (BC)', color='green', lw=2)
ax1.plot(w_values, th_pwc, label='Practical Worst Case (PWC)', color='orange', ls='--')
ax1.axvline(x=W0, color='red', ls=':', label=f'Critical WIP (W0={W0:.2f})')
ax1.set_xlabel('WIP (In-process units)')
ax1.set_ylabel('Throughput (TH)')
ax1.set_title('WIP vs Throughput')
ax1.legend()
ax1.grid(True, alpha=0.3)

# WIP vs Cycle Time (CT)
ax2.plot(w_values, ct_best, label='Best Case (BC)', color='blue', lw=2)
ax2.plot(w_values, ct_pwc, label='Practical Worst Case (PWC)', color='orange', ls='--')
ax2.axvline(x=W0, color='red', ls=':', label=f'Critical WIP (W0={W0:.2f})')
ax2.set_xlabel('WIP (In-process units)')
ax2.set_ylabel('Cycle Time (CT)')
ax2.set_title('WIP vs Cycle Time')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"瓶頸站產出率 rb = {rb} 個/小時")
print(f"若 Release Rate (R) > {rb}，系統將進入非穩態 (WIP 無限累積)。")
```

### 2.3.1. WIP 與 CT、TH 的關係
* **$TH$ (產出率)**：當 $WIP < W_0$ 時，$TH$ 隨 $WIP$ 增加而上升。但一旦 $WIP$ 超過 $W_0$ (15.375)，產出率就會受限於瓶頸站，維持在 $0.375$。此後增加再多 $WIP$ 也無法提升產能。
* **$CT$ (生產週期)**：當 $WIP < W_0$ 時，$CT$ 維持在 $T_0$ (41 小時) 附近。但當 $WIP$ 超過 $W_0$ 後，$CT$ 會開始與 $WIP$ 成正比線性增加。

```python
R_steady = 0.30     
R_unsteady = 0.40   

time_steps = np.linspace(0, 100, 101)

WIP_initial = 10

wip_steady_profile = np.full_like(time_steps, WIP_initial + 5) 
wip_unsteady_profile = WIP_initial + (R_unsteady - rb) * time_steps

```

```python
plt.figure(figsize=(10, 6))
plt.plot(time_steps, wip_steady_profile, label=f'Steady State (R={R_steady:.2f} < rb={rb:.3f})', 
         color='green', lw=2.5)
plt.plot(time_steps, wip_unsteady_profile, label=f'Unsteady State (R={R_unsteady:.2f} > rb={rb:.3f})', 
         color='red', lw=2.5, ls='--')

plt.axhline(y=WIP_initial, color='black', ls=':', label='initial WIP')
plt.fill_between(time_steps, wip_steady_profile, wip_unsteady_profile, color='red', alpha=0.1)
#plt.text(80, wip_unsteady_profile[-1]-5, 'WIP infinity accumulation', color='red', fontweight='bold')

plt.xlabel('simulation time (hours)')
plt.ylabel('WIP in the system')
plt.title('System Stability Analysis: Impact of Input Rate (R) on WIP')
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)

plt.show()

print(f"bottleneck output rate rb = {rb}")
print(f"unsteady input rate R = {R_unsteady}")
print(f"WIP accumulation rate = R - rb = {R_unsteady - rb:.4f} (units/hour)")
```

### 2.3.2 Release rate and TH
當投料速度 $R$ 超過瓶頸產出率 $r_b$ 時（如紅線所示），系統無法維持穩態，WIP 隨時間線性累積。這導致了排隊時間劇增，進而導致生產週期時間（CT）嚴重惡化。(粉色區域)  
* 綠線 (穩態)：代表系統在健康狀態下應有的 WIP 水準。
* 紅線 (非穩態)：代表在投料過快時，實際留在系統裡的 WIP。
* 粉色區域：這兩者之間的距離，就是因為 $R > r_b$ 而產生的「多餘排隊 (Excess Queue)」。   


也就是說，我們可以得出下面結論：
* **穩態 (Steady State)**：當 $R < r_b$。此時系統的產出率 $TH$ 會等於投料速度 $R$ ($TH = R$)。進多少，出多少，系統平衡。
* **非穩態 (Non-steady State)**：當 $R > r_b$。此時系統的產出率 $TH$ 會被卡死在瓶頸上限（$TH = r_b = 0.375$）。因為進度快、出貨慢，系統無法達成平衡。
