# 五階段構析 Five Stage Analytics

## Wafer Map Detective Problem - 晶圓缺陷檢測分類 問題與情境
該問題通常發生於**半導體晶圓製造產線（如蝕刻或微影製程）**。在先進製程中，晶圓圖 (Wafer Map) 的瑕疵數據往往差異極大，但有些具備特定致命瑕疵（如 Scratch 刮痕、Edge-Ring 邊緣環狀），這導致傳統的機器學習分類器難以收斂，容易產生誤報 (False Negative)，讓瑕疵品流入下一道工序，可能因此造成工廠巨大的成本損失。  
  
藉此，希望導入基於去噪分數匹配 (Denoising Score Matching) 等機制的生成式 AI 來擴增少數瑕疵類別，能平衡數據集，可以大幅提升後續預測模型的準確度與產線良率。

## Five Stage Analytics

首先奠基於**敘述性分析 (Descriptive)**。透過蒐集產線上的感測器數據與自動光學檢測 (AOI) 的晶圓圖 (Wafer Map)，系統能即時監控並具象化目前的良率狀態。當敘述性指標呈現出特定的異常分佈（例如發現極少數但致命的刮痕或邊緣環狀瑕疵）時，我們便可以沿著異常分布與警訊往成品上游找到問題製造者，以及造成這些特定圖譜的根本原因，流程自然推進至**診斷性分析 (Diagnostic)**，例如特定蝕刻機台的氣體流量失衡。  
  
找到異常造成的根本原因後，我們可以試著設計模型，根據及時串流數據提前預測一批晶圓是否有高機率報廢，這促使流程進入**預測性分析 (Predictive)**。比如我們可導入擴散模型 (Diffusion Models) 來生成高品質的瑕疵影像以擴增數據集。這份經過強化的數據被用來訓練機器學習分類器，使其能夠在機台發生微小偏移時，不僅能提前「預測」未來產出的晶圓是否面臨高報廢風險，也能洞見其異常週期。  
  
若系統偵測到該批晶圓有高機率報廢，肯定也相對應要考慮報廢晶圓的成本，比如停機作業損失與回收成本等；或是回頭向前一步驟調整參數等。因此數據流必須無縫接軌至**處方性分析 (Prescriptive)**，也就是工廠人員需設計決策出一套系統流程，使其不僅會顯示良率下降的預測警告，還會綜合考量停機檢修的成本與晶圓報廢的損失，開出最佳「處方」——例如建議立刻暫停某機台，或是建議微調特定製程參數。  
  
最後，當處方性系統的建議被驗證為高度可靠時，這條決策鏈的終極目標便是走向 **自主性分析 (Autonomous)** ────也就是系統會將人類決策者從迴圈中移除，建立先進製程控制 (APC) 的閉環系統，使得機台能夠根據預測與處方結果，即時且自主地進行參數補償與修正。而這些自動修正後的生產結果，又會產生新的數據，重新回饋到最初的「敘述性分析」階段進行監控，達成無人化、具備自我修正能力的閉環控制系統。
